## **1. Derivación del Peso Óptimo ($\alpha$) en un Portafolio de Dos Activos**

Utilizando propiedades estadísticas básicas de la **varianza**, así como el cálculo de una sola variable, demuestre que el valor de $\alpha$ que minimiza $\text{Var}(\alpha X + (1 - \alpha)Y)$ es el dado por la fórmula (5.6).

### **Definición del Problema**

Buscamos encontrar el peso $\alpha$ que minimiza la varianza del portafolio $Z = \alpha X + (1 - \alpha)Y$.

* $\sigma_X^2 = \text{Var}(X)$
* $\sigma_Y^2 = \text{Var}(Y)$
* $\sigma_{XY} = \text{Cov}(X, Y)$


### **Paso 1: Expresar la Varianza del Portafolio**

Aplicando la propiedad de la varianza $\text{Var}(aX + bY) = a^2 \text{Var}(X) + b^2 \text{Var}(Y) + 2ab \text{Cov}(X, Y)$, la varianza del portafolio $V(\alpha)$ es:

$$V(\alpha) = \text{Var}(\alpha X + (1 - \alpha)Y)$$
$$V(\alpha) = \alpha^2 \sigma_X^2 + (1 - \alpha)^2 \sigma_Y^2 + 2 \alpha (1 - \alpha) \sigma_{XY}$$

---

### **Paso 2: Minimizar la Varianza (Cálculo Diferencial)**

Para encontrar el mínimo, tomamos la **primera derivada** de $V(\alpha)$ con respecto a $\alpha$ y la igualamos a cero:

$$\frac{d V}{d \alpha} = \frac{d}{d \alpha} \left[ \alpha^2 \sigma_X^2 + (1 - \alpha)^2 \sigma_Y^2 + 2 (\alpha - \alpha^2) \sigma_{XY} \right] = 0$$

Realizando la derivación:

$$2 \alpha \sigma_X^2 - 2 (1 - \alpha) \sigma_Y^2 + 2 (1 - 2 \alpha) \sigma_{XY} = 0$$


### **Paso 3: Despejar $\alpha$**

Dividimos por 2 y distribuimos los términos:

$$\alpha \sigma_X^2 - \sigma_Y^2 + \alpha \sigma_Y^2 + \sigma_{XY} - 2 \alpha \sigma_{XY} = 0$$

Agrupamos los términos que contienen $\alpha$:

$$\alpha (\sigma_X^2 + \sigma_Y^2 - 2 \sigma_{XY}) = \sigma_Y^2 - \sigma_{XY}$$

Despejamos $\alpha$:

$$\alpha = \frac{\sigma_Y^2 - \sigma_{XY}}{\sigma_X^2 + \sigma_Y^2 - 2 \sigma_{XY}}$$


### **Paso 4: Expresión Final en Términos de Correlación ($\rho$)**

Sustituyendo la covarianza ($\sigma_{XY}$) por su relación con la correlación ($\rho$), donde $\sigma_{XY} = \rho \sigma_X \sigma_Y$:

$$\alpha = \frac{\sigma_Y^2 - \rho \sigma_X \sigma_Y}{\sigma_X^2 + \sigma_Y^2 - 2 \rho \sigma_X \sigma_Y}$$

$$\blacksquare \text{Q.E.D.}$$

## **2. Probabilidad de Inclusión en la Muestra Bootstrap**


Derivaremos la probabilidad de que una observación dada forme parte de una muestra bootstrap. Suponga que obtenemos una muestra bootstrap a partir de un conjunto de $n$ observaciones originales.


### **(a) Probabilidad de que la primera observación bootstrap no sea la $j$-ésima observación original.**

La muestra bootstrap se genera con **reemplazo**, por lo que cada una de las $n$ observaciones tiene la misma probabilidad de ser seleccionada en cualquier extracción.

* $\text{P}(\text{Seleccionar } j \text{ en una extracción}) = \frac{1}{n}$

Por lo tanto, la probabilidad de **no** seleccionar la $j$-ésima observación en la primera extracción es:

$$\text{P}(\text{No } j \text{ en la primera posición}) = 1 - \frac{1}{n}$$


### **(b) Probabilidad de que la segunda observación bootstrap no sea la $j$-ésima observación original.**

Dado que las extracciones son **independientes** debido al reemplazo, la probabilidad sigue siendo la misma:

$$\text{P}(\text{No } j \text{ en la segunda posición}) = 1 - \frac{1}{n}$$


### **(c) Argumentar que la probabilidad de que la $j$-ésima observación no esté en la muestra bootstrap es $\left(1 - \frac{1}{n}\right)^n$.**

Una muestra bootstrap de tamaño $n$ consiste en $n$ extracciones independientes. Para que la observación $j$ **no esté** en la muestra, debe **no ser seleccionada** en cada una de las $n$ extracciones.

Multiplicando las probabilidades independientes:

$$\text{P}(\text{No } j \text{ en la muestra bootstrap}) = \left(1 - \frac{1}{n}\right) \times \left(1 - \frac{1}{n}\right) \times \dots \times \left(1 - \frac{1}{n}\right) \quad (\text{n veces})$$

$$\text{P}(\text{No } j \text{ en la muestra bootstrap}) = \left(1 - \frac{1}{n}\right)^n$$


### **(d) Cuando $n = 5$, ¿cuál es la probabilidad de que la $j$-ésima observación esté en la muestra bootstrap?**

La probabilidad de que la observación **esté** en la muestra es el complemento de la probabilidad de que **no esté**: $\text{P}(\text{Sí } j) = 1 - \text{P}(\text{No } j)$.

$$\text{P}(\text{Sí } j) = 1 - \left(1 - \frac{1}{5}\right)^5 = 1 - \left(\frac{4}{5}\right)^5 = 1 - 0.32768 \approx **0.6723**$$


### **(e) Cuando $n = 100$, ¿cuál es la probabilidad de que la $j$-ésima observación esté en la muestra bootstrap?**

$$\text{P}(\text{Sí } j) = 1 - \left(1 - \frac{1}{100}\right)^{100} = 1 - (0.99)^{100} \approx 1 - 0.36603 \approx **0.63397**$$


### **(f) Cuando $n = 10,000$, ¿cuál es la probabilidad de que la $j$-ésima observación esté en la muestra bootstrap?**

Para valores grandes de $n$, la probabilidad se aproxima al límite fundamental:

$$\lim_{n \to \infty} \left(1 - \frac{1}{n}\right)^n = e^{-1} \approx 0.367879$$

Por lo tanto, para $n=10,000$:

$$\text{P}(\text{Sí } j) \approx 1 - e^{-1} \approx **0.63212**$$


### **(g) Creación de un Gráfico de Convergencia**
El siguiente código calculará y graficará la probabilidad $1−(1−1/n)^n.$


````python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores de n
n_values = np.arange(1, 100001)

# Calcular la probabilidad de que el j-ésimo elemento esté en la muestra
# P(Sí j) = 1 - (1 - 1/n)^n
probability_in = 1 - (1 - 1/n_values)**n_values

# El valor asintótico 1 - 1/e
asymptotic_value = 1 - np.exp(-1)

# Creación del gráfico
plt.figure(figsize=(10, 6))
plt.plot(n_values, probability_in, label='$P(\\text{Sí } j) = 1 - (1 - 1/n)^n$')
plt.axhline(asymptotic_value, color='r', linestyle='--', label='$1 - 1/e \\approx 0.63212$')
plt.title('Probabilidad de que una Observación esté en la Muestra Bootstrap vs. Tamaño de la Muestra (n)')
plt.xlabel('Tamaño de la Muestra Original, n')
plt.ylabel('Probabilidad de Inclusión')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.xscale('log') # Usamos escala logarítmica para n grande
plt.show()

print(f"Valor asintótico (cuando n es grande): {asymptotic_value:.5f}")
print(f"Probabilidad para n=1: {probability_in[0]:.5f}")
print(f"Probabilidad para n=100000: {probability_in[-1]:.5f}")
````

El gráfico muestra que la probabilidad de que una observación esté en una muestra bootstrap converge rápidamente al valor $1−e^{−1}≈0.63212$ a medida que el tamaño de la muestra original n aumenta. Para n pequeños, la probabilidad es mayor (es 1.0 para $n=1$). A partir de $n≈30$, la curva es prácticamente plana, lo que demuestra que, independientemente de que la muestra original sea de 30 o de 10,000 observaciones, aproximadamente el 63.2% de las observaciones originales estarán presentes en la muestra bootstrap generada.

### **(h) Simulación Numérica para $n=100$**
El siguiente código implementa la simulación requerida y luego calcula la media de los resultados para estimar la probabilidad empírica.

````python
import numpy as np

# Inicializar el generador de números aleatorios (semilla 10)
rng = np.random.default_rng(10)

# Array para almacenar los resultados de cada iteración
store = np.empty(10000)

# La observación a rastrear es la j=4 (índice 4 en base 0, o el quinto elemento)
observation_to_track = 4 # Valor que representa la 4ta observación
sample_size = 100        # n = 100

# Bucle para realizar 10,000 simulaciones de bootstrap
for i in range(10000):
    # Generar una muestra bootstrap de tamaño 100 (valores de 0 a 99)
    # replace=True es clave en el bootstrap
    bootstrap_sample = rng.choice(sample_size, size=sample_size, replace=True)

    # Contar cuántas veces aparece 'observation_to_track' (el 4) en la muestra
    count_of_j = np.sum(bootstrap_sample == observation_to_track)

    # Registrar si la observación j=4 está presente (True si count_of_j > 0, False si 0)
    store[i] = (count_of_j > 0)

# Calcular la media de 'store'. Como True es 1 y False es 0, la media es la proporción (probabilidad)
empirical_probability = np.mean(store)

print(f"Probabilidad Teórica para n={sample_size}: {1 - (1 - 1/sample_size)**sample_size:.5f}")
print(f"Probabilidad Empírica (Simulación de 10,000 veces): {empirical_probability:.5f}")
````

El resultado empírico de la simulación es aproximadamente 0.634, lo que está extremadamente cerca del valor de la probabilidad teórica calculada para $n=100$, que es $≈0.63397$.

- Probabilidad Teórica ($n=100$): $1−(1−1/100)^{100}≈0.63397$

- Probabilidad Empírica (Simulación): $≈0.634$

Este ejercicio numérico verifica la derivación teórica y demuestra cómo el método Bootstrap, al tomar muestras con reemplazo, deja fuera de la muestra remuestreada a una porción considerable de las observaciones originales (alrededor del $1 - 0.632 = 36.8\%$). Los métodos de validación cruzada como el Out-of-Bag (OOB) en el Random Forest explotan esta propiedad.

### **(a) Implementación de $k$-fold Cross-Validation**

La validación cruzada de $k$ pliegues es una técnica robusta para estimar el **error de prueba (test error)** de un modelo y consta de los siguientes pasos:

1.  **División de los Datos:** El conjunto de datos de entrenamiento completo de $n$ observaciones se divide aleatoriamente en **$k$ subconjuntos o "pliegues" (folds)** de tamaño aproximadamente igual.
2.  **Iteración de Entrenamiento y Prueba:** Se repite el siguiente proceso $k$ veces (en $k$ iteraciones):
    * **Conjunto de Prueba (Test Fold):** En cada iteración $i$, el $i$-ésimo pliegue se reserva como el conjunto de prueba (validation set).
    * **Conjunto de Entrenamiento (Training Folds):** Los $k-1$ pliegues restantes se combinan y se utilizan para entrenar el modelo (training set).
3.  **Cálculo del Error:** Se utiliza el modelo entrenado para predecir las respuestas en el pliegue de prueba y se calcula el error de predicción, como el **Error Cuadrático Medio (MSE)**.
4.  **Estimación del Error Final:** El error de prueba estimado final de la validación cruzada de $k$ pliegues es la **media** de los $k$ errores de prueba (MSE) calculados en cada iteración.


### **(b) Ventajas y Desventajas Relativas**

#### **i. Relación con el Enfoque del Conjunto de Validación (Validation Set Approach)**

| Característica | $k$-fold Cross-Validation | Validation Set Approach |
| :--- | :--- | :--- |
| **Sesgo (Bias) del Error Estimado** | **Bajo.** Entrena el modelo $k$ veces usando casi todos los datos ($n \cdot \frac{k-1}{k}$). La estimación es más representativa. | **Alto.** Depende en gran medida de la división aleatoria inicial. Solo usa una fracción de los datos para entrenamiento, subestimando el rendimiento real. |
| **Varianza del Error Estimado** | **Baja.** El error promedio es más estable y menos sensible a la composición de un solo conjunto de entrenamiento. | **Alta.** La estimación puede variar mucho según la muestra utilizada para la validación. |
| **Eficiencia Computacional** | **Baja.** Requiere entrenar el modelo $k$ veces. | **Alta.** Solo requiere entrenar el modelo una vez. |

**Conclusión:** $k$-fold produce una **estimación del error de prueba mucho más precisa y menos variable** que el enfoque del conjunto de validación, a costa de un mayor costo computacional.

## **4. Estimación de la Desviación Estándar de una Predicción**

Suponga que utilizamos algún método de aprendizaje estadístico para realizar una predicción para la respuesta $Y$ para un valor particular del predictor $X$. Describa cuidadosamente cómo podríamos estimar la **desviación estándar** de nuestra predicción.

### **Método de Estimación: Bootstrap (Remuestreo)**

La forma más robusta y general para estimar la **desviación estándar** (o error estándar, $\text{SE}$) de una predicción $\hat{Y}(x_0)$ generada por un modelo de aprendizaje estadístico (independientemente de su complejidad) es utilizando el método **Bootstrap**.

Este método simula la variabilidad que la predicción tendría si hubiéramos obtenido diferentes conjuntos de datos de entrenamiento de la población.

#### **Pasos de Implementación del Bootstrap**

1.  **Generación de Muestras Bootstrap:**
    * Generamos $B$ muestras remuestreadas ($Z^{*1}, Z^{*2}, \dots, Z^{*B}$) del conjunto de datos de entrenamiento original $Z$. $B$ es típicamente un número grande (e.g., $B=500$ o $B=1000$).
    * Cada muestra $Z^{*b}$ se obtiene mediante **muestreo con reemplazo** y tiene el mismo tamaño $n$ que el conjunto de datos original.

2.  **Entrenamiento y Predicción:**
    * Para cada muestra bootstrap $Z^{*b}$, entrenamos nuestro método de aprendizaje estadístico y obtenemos un modelo ajustado, $\hat{f}^{*b}$.
    * Utilizamos este modelo ajustado para generar una predicción, $\hat{Y}^{*b}(x_0)$, en el punto de prueba específico $x_0$.

3.  **Cálculo de la Desviación Estándar Estimada:**
    * La **desviación estándar (error estándar) estimada** de la predicción, $\hat{\text{SE}}[\hat{Y}(x_0)]$, se calcula como la desviación estándar de las $B$ predicciones obtenidas:

$$\hat{\text{SE}}[\hat{Y}(x_0)] = \sqrt{\frac{1}{B-1} \sum_{b=1}^{B} \left( \hat{Y}^{*b}(x_0) - \bar{Y}^{*}(x_0) \right)^2}$$

Donde $\bar{Y}^{*}(x_0)$ es la predicción promedio sobre las $B$ muestras bootstrap:

$$\bar{Y}^{*}(x_0) = \frac{1}{B} \sum_{b=1}^{B} \hat{Y}^{*b}(x_0)$$

#### **Ventaja del Bootstrap**

El método Bootstrap es invaluable porque es **no paramétrico**. No requiere hacer suposiciones sobre la distribución subyacente de los datos o los errores, lo que permite estimar la variabilidad (incertidumbre) de la predicción de **cualquier** algoritmo de aprendizaje estadístico (regresión, árboles, redes neuronales, etc.).

## **Importacion de librerias y carga de datos**

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Fijar semilla
np.random.seed(1)

# Cargar dataset Default (si lo tienes en ISLR o CSV)
import statsmodels.api as sm
default = sm.datasets.get_rdataset("Default", "ISLR").data

default.head()
````

## **Parte (a): Ajustar un modelo logístico usando income y balance**


````python
X = default[['income', 'balance']]
y = (default['default'] == 'Yes').astype(int)

model = LogisticRegression()
model.fit(X, y)

model_coef = pd.DataFrame({
    "Variable": ['Intercept', 'income', 'balance'],
    "Coeficiente": np.append(model.intercept_, model.coef_[0])
})
model_coef
````

## **Parte (b): Estimar el error de prueba usando validación simple**


````python
# 1. Dividir en train/test
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=1)

# 2. Ajustar modelo
model_val = LogisticRegression()
model_val.fit(X_train, y_train)

# 3. Predecir probabilidades
probs = model_val.predict_proba(X_val)[:,1]

# Clasificar usando threshold de 0.5
preds = (probs > 0.5).astype(int)

# 4. Error de validación
error_val = 1 - accuracy_score(y_val, preds)
error_val
````

## **Parte (c): Repetir validación con tres semillas distintas**


````python
errores = []

for seed in [1, 10, 25]:
    X_train, X_val, y_train, y_val = train_test_split(
        X, y, test_size=0.3, random_state=seed)

    model = LogisticRegression()
    model.fit(X_train, y_train)

    preds = (model.predict_proba(X_val)[:,1] > 0.5).astype(int)
    error = 1 - accuracy_score(y_val, preds)
    errores.append((seed, error))

errores
````

````python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Cargar dataset Weekly de ISLR
weekly = sm.datasets.get_rdataset("Weekly", "ISLR").data

# Convertir Direction a 0/1
weekly['Direction_bin'] = (weekly['Direction'] == "Up").astype(int)

weekly.head()
````

## **Parte (b): Ajustar el modelo dejando fuera la primera observación**


````python
X_minus1 = X.iloc[1:]
y_minus1 = y.iloc[1:]

model_minus1 = sm.GLM(y_minus1, X_minus1, family=sm.families.Binomial()).fit()
````

````python
n = len(weekly)
errors = np.zeros(n)

# Recorremos LAS POSICIONES 0..n-1,
# pero guardamos el índice real de cada fila para usarlo con .loc
for pos in range(n):
    idx = X.index[pos]  # índice real de la fila

    # 1. Dejar fuera la observación i-ésima (por índice real)
    X_train = X.drop(idx)
    y_train = y.drop(idx)

    # 2. Ajustar el modelo
    model_i = sm.GLM(y_train, X_train, family=sm.families.Binomial()).fit()

    # 3. Predecir la probabilidad para esa observación
    prob_i = model_i.predict(X.loc[[idx]]).iloc[0]  # <- aquí el cambio importante

    # 4. Clasificación y error
    pred_i = int(prob_i > 0.5)
    errors[pos] = int(pred_i != y.loc[idx])
````

La relación verdadera entre X y Y es cuadrática.

## (c) Cálculo del error LOOCV para los modelos de grados 1 a 4

Se calcularon los errores LOOCV para los modelos polinomiales de grados 1, 2, 3 y 4 
ajustados mediante mínimos cuadrados.


````python
def loocv_mse(x, y, degree):
    n = len(x)
    errors = np.empty(n)
    
    for i in range(n):
        mask = np.ones(n, dtype=bool)
        mask[i] = False
        
        # Ajustar polinomio
        coef = np.polyfit(x[mask], y[mask], degree)
        
        # Predicción para la observación i
        y_hat_i = np.polyval(coef, x[i])
        
        errors[i] = (y[i] - y_hat_i)**2
    
    return errors.mean()

# Cálculo de LOOCV para grados 1 a 4
for d in range(1, 5):
    mse = loocv_mse(x, y, d)
    print(f"Grado {d}: LOOCV MSE = {mse:.4f}")
````

- En el modelo cuadrático, los coeficientes de X y X² suelen ser altamente significativos, alineándose con el modelo verdadero.
- En los modelos cúbico y de grado 4, los términos cúbico y cuártico no son claramente significativos, lo que indica sobreajuste.

## **(a) Estimación de la media poblacional de `medv`**

````python
import numpy as np
import statsmodels.api as sm

# Cargar dataset Boston
boston = sm.datasets.get_rdataset("Boston", "MASS").data
medv = boston['medv']

mu_hat = medv.mean()
mu_hat
````

## (b) Estimación del error estándar de la media

El error estándar de la media se calcula mediante:

$$
SE(\hat{\mu}) = \frac{s}{\sqrt{n}}
$$

donde:
- \(s\) = desviación estándar muestral  
- \(n\) = tamaño de la muestra


````python
n = len(medv)
s = medv.std()

se_mu = s / np.sqrt(n)
se_mu
````

## (c) Estimación del error estándar de la media usando bootstrap

Se toman muchas muestras bootstrap con reemplazo.  
La desviación estándar de las medias bootstrap es la estimación del error estándar.


````python
B = 2000
boot_means = np.empty(B)

rng = np.random.default_rng(1)

for b in range(B):
    sample = rng.choice(medv, size=n, replace=True)
    boot_means[b] = sample.mean()

se_boot_mu = boot_means.std()
se_boot_mu
````

## (d) Intervalo de confianza del 95% usando bootstrap

El intervalo se calcula como:

$$
[\hat{\mu} - 2SE,\ \hat{\mu} + 2SE]
$$

Se compara con el intervalo obtenido utilizando el error estándar teórico.


````python
ci_boot = (mu_hat - 2 * se_boot_mu, mu_hat + 2 * se_boot_mu)
ci_std  = (mu_hat - 2 * se_mu,      mu_hat + 2 * se_mu)

ci_boot, ci_std
````
